```python
from youtube_transcript_api import YouTubeTranscriptApi
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import faiss
import numpy as np

# 1. Load transcript
def get_transcript(video_id):
    print("🔹 Getting transcript...")
    transcript = YouTubeTranscriptApi.get_transcript(video_id)
    full_text = " ".join([t["text"] for t in transcript])
    print("✅ Transcript fetched.")
    return full_text

# 2. Chunk the transcript
def split_text(text, chunk_size=500):
    print("🔹 Splitting text into chunks...")
    words = text.split()
    chunks = [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]
    print(f"✅ Text split into {len(chunks)} chunks.")
    return chunks

# 3. Embed using SentenceTransformer
def embed_chunks(chunks):
    print("🔹 Embedding chunks...")
    model = SentenceTransformer("all-MiniLM-L6-v2")
    embeddings = model.encode(chunks)
    print("✅ Embeddings created.")
    return embeddings

# 4. Store in FAISS
def build_faiss_index(embeddings):
    print("🔹 Building FAISS index...")
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embeddings)
    print("✅ FAISS index ready.")
    return index

# 5. Retrieve and summarize
def retrieve_and_summarize(query, chunks, embeddings, index, k=3):
    print("🔹 Retrieving relevant chunks...")
    model = SentenceTransformer("all-MiniLM-L6-v2")
    query_embedding = model.encode([query])
    _, indices = index.search(np.array(query_embedding), k)
    selected_chunks = [chunks[i] for i in indices[0]]

    print("🔹 Summarizing...")
    summarizer = pipeline("summarization", model="t5-small", tokenizer="t5-small")
    inputs = " ".join(selected_chunks)
    summary = summarizer(inputs, max_length=150, min_length=40, do_sample=False)[0]['summary_text']

    print("✅ Summary complete!")
    return summary
```

```python
from urllib.parse import urlparse, parse_qs

def extract_video_id(url):
    if "youtu.be" in url:
        return url.split("/")[-1].split("?")[0]
    elif "youtube.com" in url:
        return parse_qs(urlparse(url).query).get("v", [None])[0]
    return None
```

```python
from youtube_transcript_api._errors import TranscriptsDisabled, NoTranscriptFound

def get_transcript(video_id):
    print("🔹 Getting transcript...")
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        full_text = " ".join([t["text"] for t in transcript])
        print("✅ Transcript fetched.")
        return full_text
    except TranscriptsDisabled:
        print("❌ Subtitles are disabled for this video.")
        return None
    except NoTranscriptFound:
        print("❌ No transcript found (may be region-locked or private).")
        return None
    except Exception as e:
        print(f"❌ Unexpected error: {e}")
        return None
```

```python
video_url = "https://www.youtube.com/watch?v=2vjPBrBU-TM"  # Example: Sia - Chandelier (auto-generated subtitles)

video_id = extract_video_id(video_url)

if video_id:
    raw_text = get_transcript(video_id)

    if raw_text:
        chunks = split_text(raw_text)
        embeddings = embed_chunks(chunks)
        index = build_faiss_index(np.array(embeddings))

        query = "Summarize the main points of the video"
        summary = retrieve_and_summarize(query, chunks, embeddings, index)
        print("\n🔍 Summary:\n", summary)
    else:
        print("⚠️ Could not summarize because transcript was unavailable.")
else:
    print("❌ Invalid YouTube URL")
```
