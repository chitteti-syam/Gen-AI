```python
def fetch_article_details(id_list):
    ids = ",".join(id_list)
    handle = Entrez.esummary(db="pubmed", id=ids)
    records = Entrez.read(handle)

    articles = []
    for record in records:
        article = {
            "title": record["Title"],
            "source": record["Source"],
            "pub_date": record["PubDate"],
            "summary": record["Summary"]
        }
        articles.append(article)

    return articles
```

```python
from Bio import Entrez

Entrez.email = "your_email@example.com"  # <- Replace with your real email

def search_pubmed(query, max_results=5):
    handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results)
    record = Entrez.read(handle)
    return record["IdList"]

def fetch_article_abstracts(id_list):
    ids = ",".join(id_list)
    handle = Entrez.efetch(db="pubmed", id=ids, rettype="abstract", retmode="text")
    abstracts = handle.read()
    return abstracts
```

```python
query = "heart disease treatment"
id_list = search_pubmed(query, max_results=3)
abstracts = fetch_article_abstracts(id_list)

print("Fetched Abstracts:\n")
print(abstracts)
```
